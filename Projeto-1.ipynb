```python
import pyautogui as pa
import time

# pyautogui.click -> clicar com o mouse
# pyautogui.write -> escrever um texto
# pyautogui.press -> apertar uma tecla
# pyautogui.hotkey -> apertar uma combinação de teclas

pyautogui.PAUSE = 1

# Passo 1: Acessar o sistema
pa.hotkey("ctrl", "t")
pa.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pa.press("enter")

time.sleep(3)

# Passo 2: Fazer login no sistema
pa.click(x=685, y=311)
pa.write("meu login")

pa.click(x=669, y=387)
pa.write("minha senha")

pa.click(x=701, y=454)
time.sleep(3)

# Passo 3: Baixar base de dados
pa.click(x=419, y=411)
pa.click(x=669, y=157)
pa.click(x=717, y=497)

time.sleep(5)
```

```python
# Passo 4: Calcular os indicadores
import pandas as pd

# importar a base de indicadores
tabela = pd.read_csv(r"C:\Users\lucas\Downloads\Compras.csv", sep=";")
display(tabela)

# calculo dos indicadores
# total gasto -> somar a coluna Valor Final
total_gasto = tabela["ValorFinal"].sum()

# quantidade -> somar a coluna Quantidade
quantidade = tabela["Quantidade"].sum()

# preço medio -> total gasto / quantidade
preco_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)
```

```python
import pyperclip as pc
# Passo 5: Enviar o email para diretoria/ para o chefe

# entrar no email
pa.hotkey("ctrl", "t")
pa.write("https://mail.google.com/mail/u/0/#inbox")
pa.press("enter")

time.sleep(8)

# clicar em escrever
pa.click(x=114, y=154)

time.sleep(3)

# preencher o email
pa.write("lucasilva589@gmail.com")
pa.press("tab") # seleciona o email

pa.press("tab") # passa para o assunto
pc.copy("Relatório de Compras")
pa.hotkey("ctrl", "v")

pa.press("tab") # passar para o corpo do email

texto = f"""
Prezados,

Segue Relatório de Compras:

Total Gasto: R${total_gasto:,.2f}
Quantidade de Produtos: {quantidade:,}
Preço Médio: R${preco_medio:,.2f}

Qualquer dúvida é só falar
Att.,
Lucas Loiola
"""
pc.copy(texto)
pa.hotkey("ctrl", "v")

# enviar
pa.hotkey("ctrl", "enter")
```
